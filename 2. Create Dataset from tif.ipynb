```python
path = "../shared/data/unzip/LC080100562018022701T1-SC20191101092059"

def find_band_path(band_name,base_path):
    lookup = path + '/*.tif'
    bands_paths = glob.glob(lookup)
    
    condition = lambda file_name: band_name in file_name
    result = list(filter(condition,bands_paths))

    # result debería tener un único valor desde que 
    # un único path contiene el string que representa
    # la variable band_name
    assert len(result) == 1, 'Nombre de la banda incompleto'
    return result.pop(0)

band_names = [
    'band1','band2','band3','band4','band5','band6','band7','pixel_qa','radsat_qa','aerosol'
]

path_band_1 = find_band_path(band_name='band1',base_path=path)

dataset = rio.open(path_band_1)
# dataset
print('Shape: ', dataset.shape)

# Dataset georeferencing
# A GIS raster dataset is different from an ordinary image; its elements (or “pixels”)
# are mapped to regions on the earth’s surface. Every pixels of a dataset is contained 
# within a spatial bounding box.
print('Bounds: ', dataset.crs)

# The value of bounds attribute is derived from a more fundamental attribute: 
# the dataset’s geospatial transform.
# print('Transform:', dataset.transform)
dataset.transform
# img_folder_list = os.listdir(path)
# img_folder_list

# for img_folder in img_folder_list:
#     img_list = os.listdir("./data/"+img_folder+"/")
#     for img in img_list:
#         sat_img = xr.open_rasterio("./data/"+img_folder+"/"+img)
#         xr_img_dataset = da.to_dataset(name="img_table1")
#         show(sat_img)
```

```python
with rasterio.open(path_band_1) as dataset:

    # Read the dataset's valid data mask as a ndarray.
    mask = dataset.dataset_mask()

    # Extract feature shapes and values from the array.
    for geom, val in rasterio.features.shapes(
            mask, transform=dataset.transform):

        # Transform shapes from the dataset's own coordinate
        # reference system to CRS84 (EPSG:4326).
        geom = rasterio.warp.transform_geom(
            dataset.crs, 'EPSG:4326', geom)

        # Print GeoJSON shapes to stdout.
        print(geom)
```
