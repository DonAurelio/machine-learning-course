```python
from keras.models import load_model
model = load_model("my_model.h5")
```

```python
from dask.distributed import Client
client = Client()
client
```

```python
function = lambda args,**kwargs: np.argmax(kwargs['model'].predict([[ args ]]))
# function = lambda args,**kwargs: print(args,kwargs)

# with ProgressBar():
future = da.apply_along_axis(function,0,data_1.variable.data,dtype=np.int16, shape=(1,),model=model)
%time future.compute()
```
